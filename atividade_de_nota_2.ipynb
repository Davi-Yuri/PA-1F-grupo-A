Aprovação de aluno
Peça a nota de um aluno e informe:
Aprovado (nota ≥ 7)
Recuperação (5 ≤ nota < 7)
Reprovado (nota < 5)

```python
n = int(input("nota do aluno: "))

if n >= 7:
  print("o aluno foi aprovado")

elif n >= 5:
  print("o aluno esta de recuperação")

else:
  print("o aluno foi reprovado")
```

Classificação de idade
Receba a idade e classifique:
Criança (0–12)
Adolescente (13–17)
Adulto (18–59)
Idoso (60+)

```python
idade = int(input("qual a sua idade: "))

if idade <= 12:
  print(f"você tem {idade} anos,é uma crinça")

elif idade <= 17:
   print(f"você tem {idade} anos,é um adolecente")

elif idade <= 59:
  print(f"você tem {idade} anos,é um adulto")

else:
  print(f"você tem {idade} anos,é um idoso")
```

Desconto em compras
Peça o valor de uma compra:
≥ 100: 10% de desconto
≥ 50 e < 100: 5% de desconto
< 50: sem desconto


```python
valor = float(input("valor da compra: "))

if valor >= 100:
  print("essa compra tem 10% de desconto")

elif valor >= 50:
  print("essa compra tem 5% de desconto")

else:
  print("essa compra não tem desconto")
```
