```python
# calculadora com as 4 operações

#menu
#opção 1 - somar
#opção 2 - subtrair
#opção 3 - multiplicar
#opção 4  - divisão
#opção 5 - sair


soma = 1
sub = 2
mult = 3
div = 4
sair = 5

print("======MENU=====")
print("Olá! bem vindo a calculadora.")
print("Escolha uma opção: 1 soma,2 subtrair,3 multiplicar,4 dividir,5 sair.")

esco = ""
n1 = 0
n2 = 0

while esco != 5:
  esco = int(input("escolha uma opção: "))
  print("")

  if esco == soma:
    n1 = int(input("escolha um numero: "))

    n2 = int(input("escolha um numero: "))
    sr = n1 + n2
    print(sr)

  elif esco == sub:
    n1 = int(input("escolha um numero: "))

    n2 = int(input("escolha um numero: "))
    sbr = n1 - n2
    print(sbr)

  elif esco == mult:
    n1 = int(input("escolha um numero: "))

    n2 = int(input("escolha um numero: "))
    mr = n1 * n2
    print(mr)


  elif esco == div:
    n1 = int(input("escolha um numero: "))

    n2 = int(input("escolha um numero: "))
    dr = n1 / n2
    print(dr)


  elif esco == sair:
    print("Ok até mais!")
```

```python
# jogo inteligente

# 1 -par / impar
# 2 - positivo
#3 - maior de 02 números
#4 - sair

print("======MENU=====")
print("")

print("Opção 1 -par / impar")
print("Opção 2 - positivo")
print("Opção 3 - maior de 02 números")
print("Opção 4 - Sair")
print("")

while True:
  opçao = int(input("digite a opção desejada: "))

  if opçao == 1:
    n = int(input("digite um número: "))
    if n % 2 == 0:
      print(f"O número {n} é par!")
    else:
      print(f"o número {n} é impar")

  elif opçao == 2:
    n = int(input("digite um número: "))
    if n > 0:
      print(f"O número {n} é positivo")

    else:
      print(f"O número {n} é negativo")
```

```python
# projeto caixa eletronico

# você foi contratado para desenvolver um sistema simples de caixa eletronico

# o sistema deverá funcionar através de um menu interativo utilizando "while"

#o programa deve iniciar com um saldo inicial de R$1000,00 e permitir que o usuário escolha diderentes operações até decidir sair do sistema

#ver saldo
#depositar
#sacar dinheiro
#sair

#regras

#quando depositar somar o saldo anterior
#quando sacar subtrair do saldo anterior
#mostrar o saldo para o usuário

saldo= 1000,00

print("========MENU========")
print("Ver saldo (1)")
print("Depositar saldo (2)")
print("sacar saldo (3)")
print("Sair (4)")

esco = print(input("Escolha uma dessas opções: "))


if esco == 1:
    print(saldo)

elif esco == 2:
    deposito = float(input("quanto dinheiro você quer depositar?"));
    depsal = saldo + deposito;
    print(f"você depositou {deposito} e agora tem {depsal}");

elif esco == 3:
    sacar = float(input("quanto dinheiro você quer sacar?"));
    sasal = saldo - sacar;

    print(f"você sacou {sacar} e agora tem {sasal}");

else:
    print("Até mais")
```
