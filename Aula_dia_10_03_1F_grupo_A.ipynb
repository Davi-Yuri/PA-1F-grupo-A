fazer um programa que calcula o dobro de um número.

Entrada: um número

Processamento: Calcular o dobro

Saída: mostrar o resultado

```python
# entrada

N = int(input("escolha um numero: "))

# processamento

dobro = N * 2

#saída

print(f"o dobro de {N} é {dobro}")
```

escrever um progama que calcula a media das notas do aluno.

entrada: nome,idade, eas 04 notas do bimestre

peocessamento: calcular a media

saída:mostrar o nome do aluno e media obtida


```python
# entrada

nome = input("nome do aluno: ")
idade = int(input("idade do aluno: "))
n1 = float(input("nota do primeiro bimestre "))
n2 = float(input("nota do segundo bimestre "))
n3 = float(input("nota do terceiro bimestre "))
n4 = float(input("nota do quarto bimestre "))

# processamento

Média = (n1 + n2 + n3 + n4)/4

# saída

print(f"O aluno {nome} de {idade} anos ficou com {Média} de média")
```
