```python
# questão 1
# Crie um programa que leia a idade de uma pessoa e classifique conforme as regras abaixo:
	#	Até 17 anos: Menor de idade
	#	De 18 a 59 anos: Adulto
	#	60 anos ou mais: Idoso

nome = str(input("qual o seu nome? "))
idade = int(input("qual a sua idade? "))

if idade <= 17:
  print(f"olá {nome}, você é de menor porque tem {idade} anos")

elif idade <= 59:
  print(f" olá {nome}, você é um adulto porque tem {idade} anos")

else:
  print(f"olá {nome}, você é um idoso porque tem {idade} anos")
```

```python
# Questão 2 – Cálculo de desconto

#Crie  um programa que leia o valor de uma compra e aplique o desconto conforme as regras:
	#•	Até R$100: sem desconto
	#•	De R$101 a R$500: 10% de desconto
	#•	Acima de R$500: 20% de desconto
#Ao final, mostre o valor final da compra.

#•	Até R$100: sem desconto
valor = float(input("qual o valor? "))

if valor <= 100:
  print(f"o valor do produto é {valor}, não inclui desconto")

elif valor <= 500:
  desc1 = valor * 0.10
  resul1 = valor - desc1
  print(f"o valor do produto é {resul1}, incluindo 10% de desconto")

else:
  desc2 = valor * 0.20
  resul2 = valor - desc2
  print(f"o valor do produto é {resul2}, incluindo 20% de desconto")
```

```python
# Questão 3 – Verificação de triângulo

#Crie um programa que leia três valores e verifique se eles podem formar um triângulo.
#Para isso, a soma de dois lados deve ser maior que o terceiro.
#Mostre uma mensagem informando se é ou não um triângulo válido.

lado1 = float(input("insira o lado 1: "))
lado2 = float(input("insira o lado 2: "))
lado3 = float(input("insira o lado 3: "))

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print("o seu triangulo é valido")
else:
  print("seu triangulo NÃO é validio")
```

```python
#  Questão 4 – Sistema de energia (jogo)

#Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:
	#•	0: Morto
	#•	1 a 30: Crítico
	#•	31 a 70: Normal
	#•	71 a 100: Saudável

e = int(input("digite a energia do seu personagem: "))


if e == 0:
  print(f"Seu personagem tem {e} de energia. Ele esta morto")

elif e <= 30:
  print(f"Seu personagem tem {e} de energia. Ele esta critico")

elif e <= 70:
  print(f"Seu personagem tem {e} de energia. Ele esta normal")

else:
  print(f"Seu personagem tem {e} de energia. Ele esta saldavel")
```

```python
# Questão 5 – Média escolar

# Crie um programa que leia a média final de um aluno e classifique:
	#•	Média maior ou igual a 7: Aprovado
	#•	Média entre 5 e 6.9: Recuperação
	#•	Média de 4 pra baixo: reprovado

nome = str(input("qual o nome do aluno? "))
n1 = float(input("qual a nota 1 dele? "))
n2 = float(input("qual a nota 2 dele? "))
n3 = float(input("qual a nota 3 dele? "))
n4 = float(input("qual a nota 4 dele? "))

media = (n1 + n2 + n3 + n4) / 4

if media >= 7:
  print(f"o aluno {nome} teve {media} de média,esta aprovado")

elif media <= 5:
  print(f"o aluno {nome} teve {media} de média,esta de recuperação")

else:
  print(f"o aluno {nome} teve {media} de média,esta reprovado")
```

```python
# Questão 6 – Cálculo de salário com bônus

#Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras:
	#•	Pontuação menor que 5: sem bônus
	#•	Pontuação de 5 a 7: bônus de 5% do salário
	#•	Pontuação de 8 a 9: bônus de 10% do salário
	#•	Pontuação igual a 10: bônus de 15% do salário

#Ao final, mostre o valor do bônus e o salário final com o bônus aplicado.Média menor que 5: Reprovado

pont = int(input("pontuação do funcionario "))


if pont <= 4:
  print(f"o funcionario teve {pont} de pontos, sem bonus")

elif pont <= 7:
  print(f"o funcionario teve {pont} de pontos, 5% de bonus")

elif pont <= 9:
  print(f"o funcionario teve {pont} de pontos, 10% de bonus")

else:
  print(f"o funcionario teve {pont} de pontos, 15% de bonus")
```
