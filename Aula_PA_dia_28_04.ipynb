```python
# Escrever um programa que o usuario digita uma palavra e o programa verifica se a palavra é palindrome e imprime a palavra

p = input("escreva uma palavra: ")
invert = ""

for i in (p):

  invert = i + invert
print(invert)

if invert == p:
  print("essa palavra é palindrome")

else:
  print("essa palavra NÃO é palindrome")
```

```python
# Escrever um programa que o usuario digita um número e o programa imprime os numeros pares de 0 até o número

n = int (input("digite um numero: "))

for i in range (n):
  print(i * 2)
```
