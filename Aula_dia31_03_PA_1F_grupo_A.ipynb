```python
# Estrutura de decisôes

# Escrever um programa que recebe o nome e a idade do usuario e verifica se o usuario é crinça,adolescente e adulto

# >= 12 anos é criança
# =< 17 anos é adolescente
# Adulto

# Entrada

nome = str(input("digite o seu nome: "))
idade = int(input("digite a sua idade: "))

# processamento / Saída

if idade <= 12:
  print(f"{nome}, você tem {idade} anos de idade, você é criança.")

elif idade <= 18:
  print(f"{nome}, você tem {idade} anos de idade, você é adolecente.")

else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")
```

```python
# Em um jogo, um personagem possui uma variavel chamada energia. O programa deve:
# Se a energia for maior que que 70 - forte,se a energia estiver entre 30 e 70 - Cansado,
# se a enrgia estiver menor que 30 - fraco

energia = int(input("digite a energia do seu personagem: "))

#processamento / saída

if energia <= 30:
  print(f"Seu personagem tem {energia} de energia. Ele esta forte")

elif energia <= 70:
  print(f"Seu personagem tem {energia} de energia. Ele esta cansado")

else:
  print(f"Seu personagem tem {energia} de energia. Ele esta fraco")
```

```python
# escrever um programa que classifica os tipos de triangulo e valida se é um triangulo

# validar se os lados digitados pode formar um triangulo
# lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado > lado1

# Tipos de triangulo
# equilatero - todos os lados iguais - ==
# escaleno - todos os lados diferentes - !=
# Isoceles - dois lados iguais

lado1 = float(input("tamanho do lado 1: "))
lado2 = float(input("tamanho do lado 2: "))
lado3 = float(input("tamanho do lado 3: "))

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print("esse trinagulo é valido")

else:
  print("não é um triangulo valido")

if lado1 == lado2 == lado3 and lado1 == lado3 :
  print("seu triangulo é equilatero")

elif lado1 != lado2 != lado3 and lado1 != lado3 :
  print("seu triangulo é escaleno")

else:
  print("seu triangulo é Isoceles")
```
